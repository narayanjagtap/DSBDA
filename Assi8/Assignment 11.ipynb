```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
x14 = np.linspace(-6, 6, 30)
y14 = np.linspace(-6, 6, 30)

X1,Y1 = np.meshgrid(x14, y14)

Z1 = np.sin(np.sqrt(X1 ** 2 + Y1 ** 2))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, Y1, Z1, 50, cmap='binary')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
heart = pd.read_csv("forestfires.csv")
```

```python
heart.head()
```

```python
heart.describe()
```

```python
heart.info()
```

```python
heart.isnull().sum()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
vars_to_plot = ['day', 'DMC', 'ISI', 'wind']
sns.boxplot(data=heart[vars_to_plot])
```

```python
# Hitung IQR
Q1 = heart['DMC'].quantile(0.25)
Q3 = heart['DMC'].quantile(0.75)
IQR = Q3 - Q1

#Batas atas-batas bawah
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR

outliers = heart[(heart['DMC'] < batas_bawah) | (heart['DMC'] > batas_atas)]
heart = heart[(heart['DMC'] >= batas_bawah) & (heart['DMC'] <= batas_atas)]
```

```python
fig, ax = plt.subplots(figsize=(5,3))
sns.boxplot(heart['DMC'])
plt.show()
```

```python
#Distribusi data
for column in heart.columns:
    sns.histplot(heart[column], bins=10, kde=True, stat="density")
    plt.xlabel(column)
    plt.ylabel('Pasien')
    plt.title(f'Distribusi {column}')
    plt.show()
```

```python
Air=pd.read_csv("ForestFires.csv")
```

```python
Air.head()
```

```python
Air.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
numerical_columns = Air.select_dtypes(include=[np.number]).columns

for column in numerical_columns:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=Air[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
for column in numerical_columns:
    plt.figure(figsize=(6,3))
    sns.histplot(x=Air[column], stat="count", color="blue", bins=15, kde={'alpha': 0.5})
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
